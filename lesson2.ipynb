Вероятность события А в каждом независимом испытании 0.0015. Какова вероятность
того, что при 2000 испытаниях событие А появится 3 раза.
Для решения используем формулу Бернули и Пуасона

```python
from math import factorial
n = 2000
p = 0.0015
k = 3 # хотим проверить что событие произойдет k раз из n
def combinations(n,k):
    return factorial(n)/(factorial(k)*factorial(n-k))
# P(n,k) = c(n,k) * p ** k * q**(n-k)
def bernuli  (n,p,k):
    return combinations(n,k) * p ** k * (1-p)**(n-k)
n = 2000
p = 0.0015
k = 3
res1 = bernuli(n,p,k)
res1
```

по формуле пуасона
P(k) = L ** k / k! * e ** (-k)
L = p * n

```python
l = p * n
res2 = l**k/factorial(k)*2.72**(-k)
res2
```
